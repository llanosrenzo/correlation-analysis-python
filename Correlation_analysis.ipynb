<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">La <strong>correlaci√≥n</strong> es una medida estad√≠stica que cuantifica la relaci√≥n (lineal o mon√≥tona) entre dos variables num√©ricas. Nos indica que, cuando una variable cambia, la otra tiende a aumentar o disminuir.</p>

  <p style="color: #0057e7;">Tener en cuenta:correlaci√≥n no implica causalidad. Dos variables pueden estar correlacionadas sin que una cause a la otra. Siempre se requiere un an√°lisis m√°s profundo para inferir causalidad.</p>
</div>

# <b><span style='color:#fcc36d'>Paso 1.2 |</span><span style='color:#ff6200'> ¬øC√≥mo se interpreta la correlaci√≥n?</span></b>
‚¨ÜÔ∏è [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
    La correlaci√≥n nos permite <strong>medir el signo y magnitud de la tendencia</strong> entre dos variables.
  </p>

  <p style="color: #0057e7;">
    Tenemos dos conceptos a detallar:
  </p>

  <p style="color: #0057e7;">
    <strong>- Signo</strong>
  </p>

  <ul style="color: #0057e7; margin-left: 30px;">
    <li>Un valor <strong>positivo</strong> indica una relaci√≥n directa o positiva.</li>
    <li>Un valor <strong>negativo</strong> indica relaci√≥n indirecta, inversa o negativa.</li>
    <li>Un valor <strong>nulo</strong> indica que no existe una tendencia entre ambas variables (puede ocurrir que no exista relaci√≥n o que la relaci√≥n sea m√°s compleja, como una forma de U).</li>
  </ul>

  <p style="color: #0057e7;">
    <strong>- Magnitud (fuerza)</strong>
  </p>

  <p style="color: #0057e7;">
    La magnitud nos indica la fuerza de la relaci√≥n, y toma valores entre -1 a 1. Cuanto m√°s cercano sea el valor a los extremos del intervalo (1 o -1) m√°s fuerte ser√° la tendencia de las variables, o ser√° menor la dispersi√≥n que existe en los puntos alrededor de dicha tendencia. Cuanto m√°s cerca del cero est√© el coeficiente de correlaci√≥n, m√°s d√©bil ser√° la tendencia, es decir, habr√° m√°s dispersi√≥n en la nube de puntos.
  </p>

  <ul style="color: #0057e7; margin-left: 30px;">
    <li>si la correlaci√≥n vale 1 o -1 diremos que la correlaci√≥n es ‚Äúperfecta‚Äù,</li>
    <li>si la correlaci√≥n vale 0 diremos que las variables no est√°n correlacionadas.</li>
  </ul>

</div>

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">La covarianza mide la relaci√≥n entre dos variables aleatorias: c√≥mo una cambia con respecto a la otra. Si dos variables se mueven juntas, tienen covarianza positiva; si una aumenta mientras la otra disminuye, tienen covarianza negativa. Si no existe un patr√≥n en su relaci√≥n, la covarianza es cercana a cero.</p>

</div>

**üìå F√≥rmula de la covarianza:**

$$
\text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \mu_X)(Y_i - \mu_Y)
$$

- $X_i, Y_i$: Valores individuales de las variables  
- $\mu_X, \mu_Y$: Media de cada variable  
- $n$: Cantidad de observaciones


```python
# Covarianza del Monto del Estado de Cuenta Agosto y Septiembre 2005
x = df['BILL_AMT1'].to_numpy()
y = df['BILL_AMT2'].to_numpy()
```

```python
cov_matrix = np.cov(x,y,ddof=1)
print(cov_matrix)
```

- Se tiene, $ddof=1  \;\rightarrow \;$ se usa el denominador $n-1 \;\rightarrow \;$ **covarianza muestral**  
- Si usaras $ddof=0  \;\rightarrow \;$ se usa el denominador $n \;\rightarrow \;$ **covarianza poblacional** 

```python
df[['BILL_AMT1','BILL_AMT2']].cov()
```

# <b><span style='color:#fcc36d'>Paso 3.2 |</span><span style='color:#ff6200'> Correlaci√≥n de Pearson</span></b>
‚¨ÜÔ∏è [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
  El <strong>coeficiente de correlaci√≥n de Pearson</strong> es una medida estad√≠stica que eval√∫a la <strong>relaci√≥n lineal</strong> entre dos variables. Su valor oscila entre -1 y 1, donde -1 indica una correlaci√≥n negativa perfecta, 1 una correlaci√≥n positiva perfecta y 0 indica ausencia de correlaci√≥n.
  </p>
</div>

**üìå F√≥rmula de la correlaci√≥n de Pearson:**

$$
r_{X,Y} = \frac{\text{Cov}(X, Y)}{\sigma_X \cdot \sigma_Y}
$$

- $r_{X,Y}$: Coeficiente de correlaci√≥n de Pearson entre X y Y  
- $\text{Cov}(X, Y)$: Covarianza entre las variables X y Y  
- $\sigma_X, \sigma_Y$: Desviaci√≥n est√°ndar de X y de Y respectivamente


```python
import pandas as pd
r = df['BILL_AMT1'].corr(df['BILL_AMT2'],method = 'pearson')
print(f'Correlaci√≥n de Pearson: {r}')
```

### üìä Prueba de hip√≥tesis para la correlaci√≥n de Pearson

#### 1Ô∏è‚É£ Hip√≥tesis

- **Hip√≥tesis nula** $H_0$ :  
  $$
  \rho = 0
  $$
  No existe una correlaci√≥n lineal entre las variables $X$  y $Y$ .

- **Hip√≥tesis alternativa** $H_1$:  
  $$
  \rho \ne 0
  $$
  Existe una correlaci√≥n lineal significativa entre las variables $X$ y $Y$ .

---

#### 2Ô∏è‚É£ Estad√≠stico de prueba

El estad√≠stico de prueba $t$ para evaluar la significancia de $r$ se calcula como:

$$
t = \frac{r \sqrt{n - 2}}{\sqrt{1 - r^2}}
$$

donde:
- $r$: coeficiente de correlaci√≥n de Pearson muestral  
- $n$: tama√±o de la muestra  
- El estad√≠stico $t$ sigue una distribuci√≥n **t de Student** con $n - 2$ grados de libertad bajo $H_0$.

---

#### 3Ô∏è‚É£ Nivel de significancia

- Se utiliza com√∫nmente un nivel de significancia de:
  
  $$
  \alpha = 0.05
  $$

---

#### 4Ô∏è‚É£ Regla de decisi√≥n

- Si el valor $p$ asociado a $t$ es menor o igual a $\alpha$:  
  üëâ Se **rechaza** $ H_0$ ‚Üí Hay evidencia estad√≠stica de que **existe correlaci√≥n lineal** entre las variables.

- Si el valor $p$ es mayor que $\alpha$:  
  üëâ **No se rechaza** $H_0$ ‚Üí **No hay evidencia suficiente** para afirmar que existe correlaci√≥n lineal.

---


```python
import scipy.stats as stats
corr, p = stats.pearsonr(df['BILL_AMT1'],df['BILL_AMT2'])
print(f"r = {corr:.4f}, p-value = {p:.4f}")
```

```python
def corr_pearson(data, name_x,name_y):
    import scipy.stats as stats
    var_x = data[name_x]
    var_y = data[name_y]
    coef_pearson, p =stats.pearsonr(x=var_x,y=var_y)
    if p<= 0.05:
        print(f'r = {coef_pearson:.4f} | Se rechaza la Hipotesis Nula; es decir, existe correlaci√≥n lineal entre {name_x} y {name_y}')
    else:
        print(f'r = {coef_pearson:.4f} | No se rechaza la Hipotesis Nula; es decir, no hay evidencia suficiente que exista correlaci√≥n lineal entre {name_x} y {name_y}')
```

```python
corr_pearson(data = df,name_x='BILL_AMT1',name_y='BILL_AMT2')
```

# <b><span style='color:#fcc36d'>Paso 3.3 |</span><span style='color:#ff6200'> Correlaci√≥n de Spearman</span></b>
‚¨ÜÔ∏è [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
  La correlaci√≥n de Spearman es un coeficiente no parametrico que se utiliza cuando la relaci√≥n entre variables no es lineal o cuando los datos no tienen una distribuci√≥n normal. Tambi√©n se suele utilizar cuando hay valores at√≠picos en los datos o cuando los datos est√°n clasificados.
  <p style="color: #0057e7;">
  La correlaci√≥n de Spearman puede tomar valores entre -1 y 1, donde un valor de -1 indica una correlaci√≥n negativa perfecta, un valor de 1 indica una correlaci√≥n positiva perfecta y un valor de 0 indica que no hay correlaci√≥n entre las variables.
  </p>
</div>

**üìå F√≥rmula de la correlaci√≥n de Spearman:**

$$
r_s = 1- \frac{6\sum {d_i^2}}{n(n^2-1)}
$$
Siendo:
- $n$: La cantidad de obervaciones 
- $X_i$: El rango de obsevacion i con respecto a una variable  
- $Y_i$: El rango de obervaci√≥n i con respecto a una segunda variable. 
- $d_i = x_i - y_i$: Es la diferencia entre los rangos de $X$ y $Y$

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
    <strong>Importancia al momento de realizar la correlaci√≥n de spearman:</strong>
  </p>

  <ul style="color: #0057e7; margin-left: 30px;">
    <li> No se tiene en cuenta la magnitud de sus variables.</li>
    <li> Menos sensible a outliers</li>
    <li> No requiere normalidad</li>
    <li> Sirve para relaciones no lineales,pero si deben ser relaciones mon√≥tonas</li>
  </ul>
</div>

### üìä Prueba de hip√≥tesis para la correlaci√≥n de Spearman

#### 1Ô∏è‚É£ Hip√≥tesis

- **Hip√≥tesis nula** $H_0$ :  
  $$
  \rho_s = 0
  $$
  No existe una correlaci√≥n lineal entre las variables $X$  y $Y$ .

- **Hip√≥tesis alternativa** $H_1$:  
  $$
  \rho_s \ne 0
  $$
  Existe una correlaci√≥n lineal significativa entre las variables $X$ y $Y$ .

---

#### 2Ô∏è‚É£ Estad√≠stico de prueba

El estad√≠stico de prueba $t$ para evaluar la significancia de $r$ se calcula como:

$$
t = \frac{r_s \sqrt{n - 2}}{\sqrt{1 - r_s^2}}
$$

donde:
- $r_s$: coeficiente de correlaci√≥n de Spearman muestral  
- $n$: tama√±o de la muestra  
- El estad√≠stico $t$ sigue una distribuci√≥n **t de Student** con $n - 2$ grados de libertad bajo $H_0$.

---

#### 3Ô∏è‚É£ Nivel de significancia

- Se utiliza com√∫nmente un nivel de significancia de:
  
  $$
  \alpha = 0.05
  $$

---

#### 4Ô∏è‚É£ Regla de decisi√≥n

- Si el valor $p$ asociado a $t$ es menor o igual a $\alpha$:  
  üëâ Se **rechaza** $ H_0$ ‚Üí Hay evidencia estad√≠stica de que **existe correlaci√≥n Mon√≥tona** entre las variables.

- Si el valor $p$ es mayor que $\alpha$:  
  üëâ **No se rechaza** $H_0$ ‚Üí **No hay evidencia suficiente** para afirmar que existe correlaci√≥n mon√≥tona.

---


```python
import scipy.stats as stats

# Sup√≥n que tienes dos listas o arrays: x e y
rho, p_value = stats.spearmanr(x, y)

print(f"Spearman correlation: {rho:.3f}")
print(f"P-value: {p_value:.4f}")
```

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
    <strong>Diferencias entre la correlaci√≥n de Pearson y la de Spearman:</strong>
  </p>

  <ul style="color: #0057e7; margin-left: 30px;">
    <li>La correlaci√≥n de <strong>Pearson</strong> solo es v√°lida para relaciones <strong>lineales</strong>, mientras que la de <strong>Spearman</strong> puede aplicarse a relaciones lineales y no lineales mon√≥tonas.</li>
    <li><strong>Pearson</strong> mide la relaci√≥n en t√©rminos de <strong>similitud entre valores</strong>, mientras que <strong>Spearman</strong> se basa en el <strong>orden o ranking</strong> de los valores.</li>
    <li><strong>Pearson</strong> requiere que las variables sean <strong>cuantitativas y continuas</strong>, mientras que <strong>Spearman</strong> puede utilizarse con variables <strong>ordinales</strong> o categ√≥ricas ordinales.</li>
    <li><strong>Pearson</strong> es <strong>sensible a los valores extremos (outliers)</strong>, mientras que <strong>Spearman</strong> es m√°s robusto frente a ellos.</li>
  </ul>
</div>

# <b><span style='color:#fcc36d'>Paso 3.4 |</span><span style='color:#ff6200'> Correlaci√≥n de Kendall</span></b>
‚¨ÜÔ∏è [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
  La correlaci√≥n de Kendall: Una medida no param√©trica de asociaci√≥n de rangos
  <p style="color: #0057e7;">
  La correlaci√≥n de Kendall, tambi√©n conocida como coeficiente de concordancia de Kendall, se basa en la comparaci√≥n de los rangos de las observaciones en dos variables. Esta medida eval√∫a la similitud o concordancia entre los √≥rdenes de las observaciones en ambas variables, sin asumir una distribuci√≥n espec√≠fica.
  </p>
</div>

### üìå F√≥rmula de la correlaci√≥n de Kendall (Tau)

$$
\tau = \frac{n_c - n_d}{\frac{1}{2}n(n - 1)}
$$

**Donde:**

- $n$: n√∫mero total de observaciones  
- $n_c$: n√∫mero de pares **concordantes**  
- $n_d$: n√∫mero de pares **discordantes**  
- $ \frac{1}{2}n(n - 1) $: n√∫mero total de pares posibles sin repetici√≥n

---

### üìò Explicaci√≥n

- Un par de observaciones $ (x_i, y_i) $ y $ (x_j, y_j) $ es:
  - **Concordante** si:  
    $ (x_i - x_j)(y_i - y_j) > 0 $
  - **Discordante** si:  
    $ (x_i - x_j)(y_i - y_j) < 0 $

- La correlaci√≥n de Kendall compara cu√°ntos pares mantienen el mismo orden en ambas variables frente a los que lo contradicen.

- El coeficiente $ \tau $ var√≠a entre -1 y 1:
  - $ \tau = 1 $: asociaci√≥n mon√≥tona creciente perfecta  
  - $ \tau = -1 $: asociaci√≥n mon√≥tona decreciente perfecta  
  - $ \tau = 0 $: sin asociaci√≥n mon√≥tona

---

> üîç La correlaci√≥n de Kendall es √∫til especialmente cuando los datos son ordinales o se desea una medida robusta frente a valores at√≠picos.


```python
import scipy.stats as stats
x = df['EDUCATION']
y = df['BILL_AMT1']

tau,p_value = stats.kendalltau(x,y)
print(f"Coeficiente de Kendall Tau: = {tau:.4f}, p-value = {p_value:.4f}")
```

### üìä Prueba de hip√≥tesis para la correlaci√≥n de Kendall (Tau)

#### 1Ô∏è‚É£ Hip√≥tesis

- **Hip√≥tesis nula** $H_0 $:  
  $$
  \tau = 0
  $$
  No existe una correlaci√≥n mon√≥tona entre las variables $ X $ y $ Y $.

- **Hip√≥tesis alternativa** $ H_1 $:  
  $$
  \tau \ne 0
  $$
  Existe una correlaci√≥n mon√≥tona significativa entre las variables $ X $ y $ Y$.

---

#### 2Ô∏è‚É£ Estad√≠stico de prueba

Para muestras grandes, se usa una aproximaci√≥n mediante una distribuci√≥n normal:

$$
Z = \frac{r_k}{\sqrt{\frac{2(2n + 5)}{9n(n - 1)}}}
$$

donde:

- $r_k$: coeficiente de correlaci√≥n de Kendall muestral  
- $n $: tama√±o de la muestra  
- $ Z $: estad√≠stico que sigue una **distribuci√≥n normal est√°ndar** bajo $ H_0$

> Para muestras peque√±as, se recomienda usar el c√°lculo exacto del p-valor (que es lo que hace `scipy.stats.kendalltau` por defecto).

---

#### 3Ô∏è‚É£ Nivel de significancia

- Se utiliza com√∫nmente un nivel de significancia de:

  $$
  \alpha = 0.05
  $$

---

#### 4Ô∏è‚É£ Regla de decisi√≥n

- Si el valor $ p$ asociado al estad√≠stico $ Z$ es menor o igual a $ \alpha $:  
  üëâ Se **rechaza** $ H_0 $ ‚Üí Hay evidencia estad√≠stica de que **existe correlaci√≥n mon√≥tona** entre las variables.

- Si el valor $p$ es mayor que $\alpha $:  
  üëâ **No se rechaza** $H_0 $ ‚Üí **No hay evidencia suficiente** para afirmar que existe correlaci√≥n mon√≥tona.

---

# <b><span style='color:#fcc36d'>Paso 4.1 |</span><span style='color:#ff6200'> V de Cramer</span></b>
‚¨ÜÔ∏è [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
  La V de Cramer es una medida estad√≠stica que cuantifica la fuerza de asociaci√≥n entre dos variables categ√≥ricas o nominales. Se utiliza para analizar datos en tablas de contingencia y su valor var√≠a entre 0 y 1, donde 0 indica ausencia de asociaci√≥n y 1 indica una asociaci√≥n completa. </p>
  <p style="color: #0057e7;">
  <strong>¬øQu√© es la V de Cramer?</strong>
  </p>
  <p style="color: #0057e7;">
  La V de Cramer es una medida de asociaci√≥n que se calcula a partir del estad√≠stico chi-cuadrado de Pearson. Se utiliza para determinar la fuerza de la relaci√≥n entre dos variables categ√≥ricas, como, por ejemplo, la relaci√≥n entre el g√©nero (masculino/femenino) y la preferencia por un producto (s√≠/no). 
  </p>
</div>

### üìå F√≥rmula de la V de Cramer

La **V de Cramer** se utiliza para medir la **fuerza de asociaci√≥n entre dos variables categ√≥ricas**, basada en el estad√≠stico $ \chi^2 $ (chi-cuadrado) de una tabla de contingencia.

$$
V = \sqrt{ \frac{\chi^2}{n \cdot (k - 1)} }
$$

**Donde:**

- $ \chi^2$: valor del estad√≠stico chi-cuadrado de la tabla de contingencia  
- $ n $: n√∫mero total de observaciones  
- $ k $: el menor n√∫mero entre el n√∫mero de filas y el n√∫mero de columnas

---

### üìò Interpretaci√≥n

- $ V = 0 $: no hay asociaci√≥n entre las variables
- $ V = 1 $: asociaci√≥n perfecta (m√°xima posible)
- Valores intermedios indican distintos grados de asociaci√≥n

> ‚ö†Ô∏è A diferencia del chi-cuadrado, la **V de Cramer est√° normalizada entre 0 y 1**, lo que permite interpretar la fuerza de asociaci√≥n sin depender del tama√±o de la muestra.
