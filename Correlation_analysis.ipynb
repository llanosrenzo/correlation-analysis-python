# <b><span style='color:#fcc36d'>Paso 1.1 |</span><span style='color:#ff6200'> ¿Qué es la Correlación?</span></b>
⬆️ [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">La <strong>correlación</strong> es una medida estadística que cuantifica la relación (lineal o monótona) entre dos variables numéricas. Nos indica que, cuando una variable cambia, la otra tiende a aumentar o disminuir.</p>

  <p style="color: #0057e7;">Tener en cuenta:correlación no implica causalidad. Dos variables pueden estar correlacionadas sin que una cause a la otra. Siempre se requiere un análisis más profundo para inferir causalidad.</p>
</div>

# <b><span style='color:#fcc36d'>Paso 1.2 |</span><span style='color:#ff6200'> ¿Cómo se interpreta la correlación?</span></b>
⬆️ [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
    La correlación nos permite <strong>medir el signo y magnitud de la tendencia</strong> entre dos variables.
  </p>

  <p style="color: #0057e7;">
    Tenemos dos conceptos a detallar:
  </p>

  <p style="color: #0057e7;">
    <strong>- Signo</strong>
  </p>

  <ul style="color: #0057e7; margin-left: 30px;">
    <li>Un valor <strong>positivo</strong> indica una relación directa o positiva.</li>
    <li>Un valor <strong>negativo</strong> indica relación indirecta, inversa o negativa.</li>
    <li>Un valor <strong>nulo</strong> indica que no existe una tendencia entre ambas variables (puede ocurrir que no exista relación o que la relación sea más compleja, como una forma de U).</li>
  </ul>

  <p style="color: #0057e7;">
    <strong>- Magnitud (fuerza)</strong>
  </p>

  <p style="color: #0057e7;">
    La magnitud nos indica la fuerza de la relación, y toma valores entre -1 a 1. Cuanto más cercano sea el valor a los extremos del intervalo (1 o -1) más fuerte será la tendencia de las variables, o será menor la dispersión que existe en los puntos alrededor de dicha tendencia. Cuanto más cerca del cero esté el coeficiente de correlación, más débil será la tendencia, es decir, habrá más dispersión en la nube de puntos.
  </p>

  <ul style="color: #0057e7; margin-left: 30px;">
    <li>si la correlación vale 1 o -1 diremos que la correlación es “perfecta”,</li>
    <li>si la correlación vale 0 diremos que las variables no están correlacionadas.</li>
  </ul>

</div>

⬆️ [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">La covarianza mide la relación entre dos variables aleatorias: cómo una cambia con respecto a la otra. Si dos variables se mueven juntas, tienen covarianza positiva; si una aumenta mientras la otra disminuye, tienen covarianza negativa. Si no existe un patrón en su relación, la covarianza es cercana a cero.</p>

</div>

**📌 Fórmula de la covarianza:**

$$
\text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \mu_X)(Y_i - \mu_Y)
$$

- $X_i, Y_i$: Valores individuales de las variables  
- $\mu_X, \mu_Y$: Media de cada variable  
- $n$: Cantidad de observaciones


```python
# Covarianza del Monto del Estado de Cuenta Agosto y Septiembre 2005
x = df['BILL_AMT1'].to_numpy()
y = df['BILL_AMT2'].to_numpy()
```

```python
cov_matrix = np.cov(x,y,ddof=1)
print(cov_matrix)
```

- Se tiene, $ddof=1  \;\rightarrow \;$ se usa el denominador $n-1 \;\rightarrow \;$ **covarianza muestral**  
- Si usaras $ddof=0  \;\rightarrow \;$ se usa el denominador $n \;\rightarrow \;$ **covarianza poblacional** 

```python
df[['BILL_AMT1','BILL_AMT2']].cov()
```

# <b><span style='color:#fcc36d'>Paso 3.2 |</span><span style='color:#ff6200'> Correlación de Pearson</span></b>
⬆️ [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
  El <strong>coeficiente de correlación de Pearson</strong> es una medida estadística que evalúa la <strong>relación lineal</strong> entre dos variables. Su valor oscila entre -1 y 1, donde -1 indica una correlación negativa perfecta, 1 una correlación positiva perfecta y 0 indica ausencia de correlación.
  </p>
</div>

**📌 Fórmula de la correlación de Pearson:**

$$
r_{X,Y} = \frac{\text{Cov}(X, Y)}{\sigma_X \cdot \sigma_Y}
$$

- $r_{X,Y}$: Coeficiente de correlación de Pearson entre X y Y  
- $\text{Cov}(X, Y)$: Covarianza entre las variables X y Y  
- $\sigma_X, \sigma_Y$: Desviación estándar de X y de Y respectivamente


```python
import pandas as pd
r = df['BILL_AMT1'].corr(df['BILL_AMT2'],method = 'pearson')
print(f'Correlación de Pearson: {r}')
```

### 📊 Prueba de hipótesis para la correlación de Pearson

#### 1️⃣ Hipótesis

- **Hipótesis nula** $H_0$ :  
  $$
  \rho = 0
  $$
  No existe una correlación lineal entre las variables $X$  y $Y$ .

- **Hipótesis alternativa** $H_1$:  
  $$
  \rho \ne 0
  $$
  Existe una correlación lineal significativa entre las variables $X$ y $Y$ .

---

#### 2️⃣ Estadístico de prueba

El estadístico de prueba $t$ para evaluar la significancia de $r$ se calcula como:

$$
t = \frac{r \sqrt{n - 2}}{\sqrt{1 - r^2}}
$$

donde:
- $r$: coeficiente de correlación de Pearson muestral  
- $n$: tamaño de la muestra  
- El estadístico $t$ sigue una distribución **t de Student** con $n - 2$ grados de libertad bajo $H_0$.

---

#### 3️⃣ Nivel de significancia

- Se utiliza comúnmente un nivel de significancia de:
  
  $$
  \alpha = 0.05
  $$

---

#### 4️⃣ Regla de decisión

- Si el valor $p$ asociado a $t$ es menor o igual a $\alpha$:  
  👉 Se **rechaza** $ H_0$ → Hay evidencia estadística de que **existe correlación lineal** entre las variables.

- Si el valor $p$ es mayor que $\alpha$:  
  👉 **No se rechaza** $H_0$ → **No hay evidencia suficiente** para afirmar que existe correlación lineal.

---


```python
import scipy.stats as stats
corr, p = stats.pearsonr(df['BILL_AMT1'],df['BILL_AMT2'])
print(f"r = {corr:.4f}, p-value = {p:.4f}")
```

```python
def corr_pearson(data, name_x,name_y):
    import scipy.stats as stats
    var_x = data[name_x]
    var_y = data[name_y]
    coef_pearson, p =stats.pearsonr(x=var_x,y=var_y)
    if p<= 0.05:
        print(f'r = {coef_pearson:.4f} | Se rechaza la Hipotesis Nula; es decir, existe correlación lineal entre {name_x} y {name_y}')
    else:
        print(f'r = {coef_pearson:.4f} | No se rechaza la Hipotesis Nula; es decir, no hay evidencia suficiente que exista correlación lineal entre {name_x} y {name_y}')
```

```python
corr_pearson(data = df,name_x='BILL_AMT1',name_y='BILL_AMT2')
```

# <b><span style='color:#fcc36d'>Paso 3.3 |</span><span style='color:#ff6200'> Correlación de Spearman</span></b>
⬆️ [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
  La correlación de Spearman es un coeficiente no parametrico que se utiliza cuando la relación entre variables no es lineal o cuando los datos no tienen una distribución normal. También se suele utilizar cuando hay valores atípicos en los datos o cuando los datos están clasificados.
  <p style="color: #0057e7;">
  La correlación de Spearman puede tomar valores entre -1 y 1, donde un valor de -1 indica una correlación negativa perfecta, un valor de 1 indica una correlación positiva perfecta y un valor de 0 indica que no hay correlación entre las variables.
  </p>
</div>

**📌 Fórmula de la correlación de Spearman:**

$$
r_s = 1- \frac{6\sum {d_i^2}}{n(n^2-1)}
$$
Siendo:
- $n$: La cantidad de obervaciones 
- $X_i$: El rango de obsevacion i con respecto a una variable  
- $Y_i$: El rango de obervación i con respecto a una segunda variable. 
- $d_i = x_i - y_i$: Es la diferencia entre los rangos de $X$ y $Y$

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
    <strong>Importancia al momento de realizar la correlación de spearman:</strong>
  </p>

  <ul style="color: #0057e7; margin-left: 30px;">
    <li> No se tiene en cuenta la magnitud de sus variables.</li>
    <li> Menos sensible a outliers</li>
    <li> No requiere normalidad</li>
    <li> Sirve para relaciones no lineales,pero si deben ser relaciones monótonas</li>
  </ul>
</div>

### 📊 Prueba de hipótesis para la correlación de Spearman

#### 1️⃣ Hipótesis

- **Hipótesis nula** $H_0$ :  
  $$
  \rho_s = 0
  $$
  No existe una correlación lineal entre las variables $X$  y $Y$ .

- **Hipótesis alternativa** $H_1$:  
  $$
  \rho_s \ne 0
  $$
  Existe una correlación lineal significativa entre las variables $X$ y $Y$ .

---

#### 2️⃣ Estadístico de prueba

El estadístico de prueba $t$ para evaluar la significancia de $r$ se calcula como:

$$
t = \frac{r_s \sqrt{n - 2}}{\sqrt{1 - r_s^2}}
$$

donde:
- $r_s$: coeficiente de correlación de Spearman muestral  
- $n$: tamaño de la muestra  
- El estadístico $t$ sigue una distribución **t de Student** con $n - 2$ grados de libertad bajo $H_0$.

---

#### 3️⃣ Nivel de significancia

- Se utiliza comúnmente un nivel de significancia de:
  
  $$
  \alpha = 0.05
  $$

---

#### 4️⃣ Regla de decisión

- Si el valor $p$ asociado a $t$ es menor o igual a $\alpha$:  
  👉 Se **rechaza** $ H_0$ → Hay evidencia estadística de que **existe correlación Monótona** entre las variables.

- Si el valor $p$ es mayor que $\alpha$:  
  👉 **No se rechaza** $H_0$ → **No hay evidencia suficiente** para afirmar que existe correlación monótona.

---


```python
import scipy.stats as stats

# Supón que tienes dos listas o arrays: x e y
rho, p_value = stats.spearmanr(x, y)

print(f"Spearman correlation: {rho:.3f}")
print(f"P-value: {p_value:.4f}")
```

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
    <strong>Diferencias entre la correlación de Pearson y la de Spearman:</strong>
  </p>

  <ul style="color: #0057e7; margin-left: 30px;">
    <li>La correlación de <strong>Pearson</strong> solo es válida para relaciones <strong>lineales</strong>, mientras que la de <strong>Spearman</strong> puede aplicarse a relaciones lineales y no lineales monótonas.</li>
    <li><strong>Pearson</strong> mide la relación en términos de <strong>similitud entre valores</strong>, mientras que <strong>Spearman</strong> se basa en el <strong>orden o ranking</strong> de los valores.</li>
    <li><strong>Pearson</strong> requiere que las variables sean <strong>cuantitativas y continuas</strong>, mientras que <strong>Spearman</strong> puede utilizarse con variables <strong>ordinales</strong> o categóricas ordinales.</li>
    <li><strong>Pearson</strong> es <strong>sensible a los valores extremos (outliers)</strong>, mientras que <strong>Spearman</strong> es más robusto frente a ellos.</li>
  </ul>
</div>

# <b><span style='color:#fcc36d'>Paso 3.4 |</span><span style='color:#ff6200'> Correlación de Kendall</span></b>
⬆️ [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
  La correlación de Kendall: Una medida no paramétrica de asociación de rangos
  <p style="color: #0057e7;">
  La correlación de Kendall, también conocida como coeficiente de concordancia de Kendall, se basa en la comparación de los rangos de las observaciones en dos variables. Esta medida evalúa la similitud o concordancia entre los órdenes de las observaciones en ambas variables, sin asumir una distribución específica.
  </p>
</div>

### 📌 Fórmula de la correlación de Kendall (Tau)

$$
\tau = \frac{n_c - n_d}{\frac{1}{2}n(n - 1)}
$$

**Donde:**

- $n$: número total de observaciones  
- $n_c$: número de pares **concordantes**  
- $n_d$: número de pares **discordantes**  
- $ \frac{1}{2}n(n - 1) $: número total de pares posibles sin repetición

---

### 📘 Explicación

- Un par de observaciones $ (x_i, y_i) $ y $ (x_j, y_j) $ es:
  - **Concordante** si:  
    $ (x_i - x_j)(y_i - y_j) > 0 $
  - **Discordante** si:  
    $ (x_i - x_j)(y_i - y_j) < 0 $

- La correlación de Kendall compara cuántos pares mantienen el mismo orden en ambas variables frente a los que lo contradicen.

- El coeficiente $ \tau $ varía entre -1 y 1:
  - $ \tau = 1 $: asociación monótona creciente perfecta  
  - $ \tau = -1 $: asociación monótona decreciente perfecta  
  - $ \tau = 0 $: sin asociación monótona

---

> 🔍 La correlación de Kendall es útil especialmente cuando los datos son ordinales o se desea una medida robusta frente a valores atípicos.


```python
import scipy.stats as stats
x = df['EDUCATION']
y = df['BILL_AMT1']

tau,p_value = stats.kendalltau(x,y)
print(f"Coeficiente de Kendall Tau: = {tau:.4f}, p-value = {p_value:.4f}")
```

### 📊 Prueba de hipótesis para la correlación de Kendall (Tau)

#### 1️⃣ Hipótesis

- **Hipótesis nula** $H_0 $:  
  $$
  \tau = 0
  $$
  No existe una correlación monótona entre las variables $ X $ y $ Y $.

- **Hipótesis alternativa** $ H_1 $:  
  $$
  \tau \ne 0
  $$
  Existe una correlación monótona significativa entre las variables $ X $ y $ Y$.

---

#### 2️⃣ Estadístico de prueba

Para muestras grandes, se usa una aproximación mediante una distribución normal:

$$
Z = \frac{r_k}{\sqrt{\frac{2(2n + 5)}{9n(n - 1)}}}
$$

donde:

- $r_k$: coeficiente de correlación de Kendall muestral  
- $n $: tamaño de la muestra  
- $ Z $: estadístico que sigue una **distribución normal estándar** bajo $ H_0$

> Para muestras pequeñas, se recomienda usar el cálculo exacto del p-valor (que es lo que hace `scipy.stats.kendalltau` por defecto).

---

#### 3️⃣ Nivel de significancia

- Se utiliza comúnmente un nivel de significancia de:

  $$
  \alpha = 0.05
  $$

---

#### 4️⃣ Regla de decisión

- Si el valor $ p$ asociado al estadístico $ Z$ es menor o igual a $ \alpha $:  
  👉 Se **rechaza** $ H_0 $ → Hay evidencia estadística de que **existe correlación monótona** entre las variables.

- Si el valor $p$ es mayor que $\alpha $:  
  👉 **No se rechaza** $H_0 $ → **No hay evidencia suficiente** para afirmar que existe correlación monótona.

---

# <b><span style='color:#fcc36d'>Paso 4.1 |</span><span style='color:#ff6200'> V de Cramer</span></b>
⬆️ [Tabla de Contenidos](#contents_table_correlation)

<div style="width: 90%; background-color: #ffffff; color: #000000; padding: 20px; border: 3px solid #d62d20; margin-bottom: 20px; border-radius: 10px; font-size: 16px;">
  <p style="color: #0057e7;">
  La V de Cramer es una medida estadística que cuantifica la fuerza de asociación entre dos variables categóricas o nominales. Se utiliza para analizar datos en tablas de contingencia y su valor varía entre 0 y 1, donde 0 indica ausencia de asociación y 1 indica una asociación completa. </p>
  <p style="color: #0057e7;">
  <strong>¿Qué es la V de Cramer?</strong>
  </p>
  <p style="color: #0057e7;">
  La V de Cramer es una medida de asociación que se calcula a partir del estadístico chi-cuadrado de Pearson. Se utiliza para determinar la fuerza de la relación entre dos variables categóricas, como, por ejemplo, la relación entre el género (masculino/femenino) y la preferencia por un producto (sí/no). 
  </p>
</div>

### 📌 Fórmula de la V de Cramer

La **V de Cramer** se utiliza para medir la **fuerza de asociación entre dos variables categóricas**, basada en el estadístico $ \chi^2 $ (chi-cuadrado) de una tabla de contingencia.

$$
V = \sqrt{ \frac{\chi^2}{n \cdot (k - 1)} }
$$

**Donde:**

- $ \chi^2$: valor del estadístico chi-cuadrado de la tabla de contingencia  
- $ n $: número total de observaciones  
- $ k $: el menor número entre el número de filas y el número de columnas

---

### 📘 Interpretación

- $ V = 0 $: no hay asociación entre las variables
- $ V = 1 $: asociación perfecta (máxima posible)
- Valores intermedios indican distintos grados de asociación

> ⚠️ A diferencia del chi-cuadrado, la **V de Cramer está normalizada entre 0 y 1**, lo que permite interpretar la fuerza de asociación sin depender del tamaño de la muestra.
